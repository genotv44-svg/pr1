```python

def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    step = 1

    while left <= right:
        print(f"Крок {step}: шукаємо між індексами {left} та {right}")
        mid = (left + right) // 2
        print(f"Середній індекс: {mid}, значення: {arr[mid]}")

        if arr[mid] == target:
            print(f"Знайдено число {target} на індексі {mid}!")
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

        step += 1

    print(f"Число {target} відсутнє в списку.")
    return -1


input_list = input("Введіть список чисел через кому: ")
numbers = [int(x.strip()) for x in input_list.split(",")]

target = int(input("Введіть число для пошуку: "))

numbers.sort()
print(f"Відсортований список: {numbers}")

binary_search(numbers, target)
```
